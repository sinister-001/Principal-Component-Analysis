### Finding optimal number of Principal Components

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=None)
pca.fit(temp_df)

exp_var = pca.explained_variance_ratio_ * 100
cum_exp_var = np.cumsum(exp_var)

plt.figure(figsize=(20, 12))

plt.bar(range(1, 129), exp_var, align='center',
        label='Individual explained variance')

plt.step(range(1, 129), cum_exp_var, where='mid',
         label='Cumulative explained variance', color='red')

plt.ylabel('Explained variance percentage')
plt.xlabel('Principal component index')
li = list(range(1, 129))
plt.xticks(ticks=li, rotation=45)
plt.legend(loc='best')
plt.tight_layout()

plt.savefig("Barplot.png")

elbow_index = np.where(cum_exp_var >= 95)[0][0]

optimal_num_components = elbow_index + 1
print("Optimal number of components:", optimal_num_components)
```

### Modelling of the data

```python
class BayesClassifier:
    def __init__(self, X_train, y_train):
        self.X_train = X_train
        self.y_train = y_train
        self.classes = y_train.unique()
        self.apriori = {}
        self.mean = {}
        self.cov = {}
        self.inv_cov = {}
        self.det_cov = {}
        
    def fit(self):
        for c in self.classes:
            X_c = self.X_train[self.y_train == c]
            self.apriori[c] = len(X_c) / len(X_train)
            self.mean[c] = np.mean(X_c, axis=0)
            self.cov[c] = covariance(X_c)
            self.inv_cov[c] = np.linalg.inv(self.cov[c])
            self.det_cov[c] = np.linalg.det(self.cov[c])
    
    def predict(self, X_test):
        y_pred = []
        for x in X_test.values:
            probs = [self.calc_prob(x, c) for c in self.classes]
            y_pred.append(self.classes[np.argmax(probs)])
        return y_pred
    
    def calc_prob(self, x, c):
        mean = self.mean[c]
        inv_cov = self.inv_cov[c]
        det_cov = self.det_cov[c]
        prob = np.exp(np.dot(-0.5 * (x - mean).T, inv_cov).dot(x - mean)) / np.power((2 * np.pi) ** len(x) * det_cov, 0.5)
        return prob * self.apriori[c]
```

```python
def covariance(X):
    X = X - np.mean(X, axis=0)
    return np.dot(X.T, X) / (len(X) - 1)
```

```python
def split_train_test(df, y_col, num_test_rows):
    # Separate the DataFrame into 'male' and 'female' classes
    male_df = df[df[y_col] == 'male']
    female_df = df[df[y_col] == 'female']

    # Take the first `num_test_rows` rows from each class for testing
    test_data = pd.concat([male_df[:num_test_rows], female_df[:num_test_rows]], ignore_index=True)

    # The remaining rows go into the training set
    train_data = pd.concat([male_df[num_test_rows:], female_df[num_test_rows:]], ignore_index=True)

    # Separate features (X) and target (y) columns
    X_train, y_train = train_data.drop(y_col, axis=1), train_data[y_col]
    X_test, y_test = test_data.drop(y_col, axis=1), test_data[y_col]

    return X_train, y_train, X_test, y_test
```

```python
def accuracy(y_true, y_pred):
    return np.sum(y_true == y_pred) / len(y_true)
```

```python
def accuracy_by_class(y_true, y_pred):
    classes = y_true.unique()
    acc_by_class = {}
    for c in classes:
        y_true_c = y_true[y_true == c]
        y_pred_c = y_pred[y_true.index[y_true == c]]
        acc_by_class[c] = accuracy(y_true_c, y_pred_c)
    return acc_by_class
```

```python
X_train, Y_train, X_test, Y_test = split_train_test(final_df, 'Class', 10)
print(f"X_train: {len(X_train)}")
print(f"X_test: {len(X_test)}")
```

```python
bayes_classifier = BayesClassifier(X_train, Y_train)
bayes_classifier.fit()
```

```python
Y_preds = bayes_classifier.predict(X_test)
result = pd.DataFrame({'Actual Class': Y_test, 'Predicted Class': Y_preds})
result
```

```python
acc = accuracy(Y_test, Y_preds)
acc_by_class = accuracy_by_class(Y_test, pd.Series(Y_preds))
print(f"Accuracy: {acc}")
print(f"Accuracy by class: {acc_by_class}")
```

```python
all_acc = {**acc_by_class, 'Overall': acc}

plt.figure(figsize=(10, 6))

bars = plt.bar(all_acc.keys(), all_acc.values())

for bar in bars:
    height = bar.get_height()
    plt.annotate(f'{height:.2f}', xy=(bar.get_x() + bar.get_width() / 2, height), xytext=(0, 3),
                 textcoords='offset points', ha='center', va='bottom')

plt.xlabel('Class')
plt.ylabel('Accuracy')
plt.title('Class wise Accuracy')
plt.show()
```
